```python
x = np.linspace(0 , np.pi , 10000)


layout = go.Layout(
    title = 'Example',
    yaxis = dict(
        title = 'volts'
    ),
    xaxis = dict(
        title = 'NanoSeconds'
    ),
)

trace1 = go.Scatter(
    y =  x*(x)+6,
    mode='lines',
    name = 'sin(x)',
    line = dict(
        shape = 'spline',
    )
   
)

fig = go.Figure(data = [trace1] , layout = layout)

py.offline.iplot(fig)
```
